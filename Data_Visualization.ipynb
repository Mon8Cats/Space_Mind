```python
plt.style.use('seaborn-whitegrid')
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.r');
```

```python
#from sklearn.gaussian_process import GaussianProcess
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
model = lambda x: x*np.sin(x)
xdata = np.array([1,3,5,6,8])
ydata = model(xdata)

#gp = GaussianProcessRegressor(corr='cubic', theta0=1e2, thetaL=1e-4, thetaU=1e1, random_start=100)
kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0,10, 1000)
yfit, sigma = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit = 2*np.sqrt(sigma)
```

```python
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit-dyfit, yfit+dyfit, color='gray', alpha=0.2)
plt.xlim(0, 10);
```

```python
plt.style.use('seaborn-white')
def f(x, y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X, Y = np.meshgrid(x, y)
%print(X.shape)
%print(Y.shape)
Z = f(X, Y)
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy'); # red gray
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy') # fill contour
plt.colorbar();
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8) # contour label
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar();
```
